```python
import pygame
import math
import random

pygame.init()
width, height = 640,480
screen = pygame.display.set_mode((width, height))
acc = [0,0]
arrows = []
badtimer = 100
badtimer1 = 0
badguys = [[640,100]]
healthvalue = 200

player = pygame.image.load("resources/images/dude.png")
grass = pygame.image.load("resources/images/grass.png")
castle = pygame.image.load("resources/images/castle.png")
arrow = pygame.image.load("resources/images/bullet.png")
badguyimg = pygame.image.load("resources/images/badguy.png")


keys = [False, False, False, False]
playerpos = [100,100]

running = True
while running:
    
    # 화면을 깨끗하게 한다. (0,0,0) 검은색
    screen.fill((0,0,0))
    
    
    #이미지를 나타낸다
    for x in range(width//grass.get_width()+1):
        for y in range(height//grass.get_height()+1):
            screen.blit(grass, (x*100,y*100))
    
    # 성 출력
    for i in range(4):
        screen.blit(castle,(0,112*i+20))
    
    # 캐릭터 회전
    position = pygame.mouse.get_pos()
    angle = math.atan2(position[1] - (playerpos[1]+32), position[0]-(playerpos[0]+26))
    playerrot = pygame.transform.rotate(player, 360-angle*57.29)
    playerposfix = (playerpos[0]-playerrot.get_rect().width//2, playerpos[1]-playerrot.get_rect().height//2)
    screen.blit(playerrot,playerposfix)
    
    # 화살
    for bullet in arrows:
        index = 0
        velx = math.cos(bullet[0])*3
        vely = math.sin(bullet[0])*3
        bullet[1] += velx
        bullet[2] += vely
        if bullet[1]<-64 or bullet[1]>640 or bullet[2]<-64 or bullet[2]>480:
            arrows.pop(index)
        index+=1
    
    for shoot in arrows:
        arrowrot = pygame.transform.rotate(arrow, 360-shoot[0]*57.29)
        screen.blit(arrowrot, (shoot[1] , shoot[2]))
    
    # 적 출현 시간 계산
    badtimer -= 1
    if badtimer == 0:
        badguys.append([640, random.randint(50,430)])
        badtimer = 100 - (badtimer1*2)
        if badtimer1>=35:
            badtimer1=35
        else:
            badtimer1+=5
    
    # 적 출현
    index=0
    for badguy in badguys:
        badguy[0]-=3
        badrect = pygame.Rect(badguyimg.get_rect())
        badrect.top = badguy[1]
        badrect.left= badguy[0]
        if badrect.left<64:
            healthvalue -= random.randint(5,20)
            badguys.pop(index)
            
    # 다음 적 출현
        index+=1
    for badguy in badguys:
        screen.blit(badguyimg, badguy)
    
     
    
    # 화면을 다시 그린다. 
    pygame.display.flip()
    
    
    # 이벤트
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_w:
                keys[0] = True
            elif event.key == pygame.K_a:
                keys[1] = True
            elif event.key == pygame.K_s:
                keys[2] = True
            elif event.key == pygame.K_d:
                keys[3] = True
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_w:
                keys[0] = False
            elif event.key == pygame.K_a:
                keys[1] = False
            elif event.key == pygame.K_s:
                keys[2] = False
            elif event.key == pygame.K_d:
                keys[3] = False
        
        # 마우스 클릭시 화살 나가게
        if event.type == pygame.MOUSEBUTTONDOWN:
            position = pygame.mouse.get_pos()
            acc[1] = acc[1] + 1
            arrows.append([math.atan2(position[1]-(playerposfix[1]+32), \
                                      position[0]-(playerposfix[0]+26)), playerposfix[0]+26, playerposfix[1]+32])
        
            
    # 키보드 방향키 조절
    if keys[0]:
        playerpos[1] = playerpos[1] - 2
    if keys[2]:
        playerpos[1] = playerpos[1] + 2
    if keys[1]:
        playerpos[0] = playerpos[0] - 2
    if keys[3]:
        playerpos[0] = playerpos[0] + 2
            
            
```
